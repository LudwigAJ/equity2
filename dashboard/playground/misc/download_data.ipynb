```python
# Import Dependencies #
#---------------------#

import yfinance as yf

import pandas as pd
import numpy as np
```

```python
# Configure the ETFs to pull #
#----------------------------#

ETF_NASDAQ_UCITS = [
    'EQQQ.L', # NASDAQ100
    'LQQ.PA', # 2x NASDAQ100
    'QQQ3.L' # 3x NASDAQ100
]

ETF_FTSE_UCITS = [
    '^FTSE', 
    'ISF.L', # 1x FTSE 
    'LUK2.L', # 2x FTSE
    '3UKL.L' # 3x FTSE
]
```

```python
dfFTSERaw = yf.Tickers(ETF_FTSE_UCITS).history(period='10y')

dfFTSERaw
```

```python
dfFTSE = dfFTSERaw.melt(ignore_index=False)
dfFTSE.columns = ['metric', 'ticker', 'value']

%cd /Users/ludwigjonsson/Documents/contabo/equity2/dashboard/playground/misc/
dfFTSE.to_csv('ftse.csv')
```
